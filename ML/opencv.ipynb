```python
import cv2
import matplotlib.pyplot as plt
```

```python
y = cv2.imread("/content/1245462.jpg",cv2.IMREAD_COLOR)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(y)
```

```python
rgb_img = cv2.imread('/content/1245462.jpg',cv2.IMREAD_GRAYSCALE)
cv2_imshow(rgb_img)
```

IMG PROPERTIES

```python
h,w,channel = y.shape
print(f'Dimensions{h}*{w}')
print(f'maximum pixel value {y.max()}')
print(f'minimum pixel value {y.min()}')
print(f'channels{channel}')
print(f'mean pixel values {y.mean()}')
```

```python
y.shape
```

```python
resize_img = cv2.resize(y,(1000,800))
cv2_imshow(resize_img)
```

```python
blurr = cv2.blur(y,(15,15))
cv2_imshow(blurr)
```

```python
medianblur = cv2.medianBlur(y,15)
cv2_imshow(medianblur)
```

```python
gauss = cv2.GaussianBlur(y,(5,5),0)
cv2_imshow(gauss)
```

sobel


```python
sobelx = cv2.Sobel(y,cv2.CV_64F,1,0,ksize=3)
cv2_imshow(sobelx)
```

```python
sobely = cv2.Sobel(y,cv2.CV_64F,0,1,ksize=3)
cv2_imshow(sobely)
```
